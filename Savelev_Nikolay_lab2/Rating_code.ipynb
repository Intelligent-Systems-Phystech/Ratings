```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize_scalar
from matplotlib import pyplot as plt
from scipy.stats.stats import pearsonr
```

```python
A = pd.read_csv('A.csv', header=None)
```

```python
Psi = pd.read_csv('Psi.csv')
```

```python
A = np.array(A.values, dtype=float)
```

```python
P = np.array(Psi.values[:, 1:], dtype = float)
```

```python
w0 = P.T
```

```python
q0 = A[:, 5:].T
```

```python
q1 = (A@(w0.T)).T
```

```python
A_pinv = np.linalg.pinv(A)
```

```python
w1 = (A_pinv@(q0.T)).T
```

```python
n = q0.shape[1]
m = w0.shape[1]
```

```python
def find_alpha(q0, q1, w0, w1):
    def f(a):
        q = a*q0 + (1-a)*q1
        w = (1-a)*w0 + a*w1
        return (np.linalg.norm(q-q0)**2)/n + (np.linalg.norm(w-w0)**2)/m
    return minimize_scalar(f, bounds=(0, 1), method='Bounded').x
```

```python
referees = A[:,5:20].T
```

```python
distances = np.zeros((15, 15))

for i in range(15):
    for j in range(15):
        ans = 0
        for k in range(9):
            if referees[i][k] != 0 and referees[j][k] != 0:
                ans += (referees[i][k]-referees[j][k])**2
        
        distances[i,j] = ans
        
        distances[i,i] = 10**6
    

for i in range(15):
    d = [(c, i) for i, c in enumerate(distances[i])]
    d.sort()
    for k in range(9):
        if referees[i][k] == 0:
            ans = 0
            n = 0
            for c, j in d:
                if referees[j][k] != 0 and n < 3:
                    ans += referees[j][k]
                    n += 1
                referees[i][k] = ans/3
```

```python
R = np.zeros((10,10))
for i in range(9):
    for j in range(i,9):
        R[i,j] = pearsonr(A[i,5:], A[j,5:])[0]
        
G = np.zeros((20,20))
for i in range(20):
    for j in range(i,20):
        G[i,j] = pearsonr(w0[:,i], w0[:,j])[0]
```

```python
def find_alpha2(q0, q1, w0, w1):
    def f(a):
        def distQ(x,y):
            ans = R.dot(x-y)
            return (x-y).dot(ans)
        def distW(x,y):
            ans = G.dot(x-y)
            return (x-y).dot(ans)
        q = a*q0 + (1-a)*q1
        w = (1-a)*w0 + a*w1
        return 1/n*distQ(q,q0) + 1/m*distW(w,w0)
    return minimize_scalar(f, bounds=(0, 1), method='Bounded').x
```

```python
rangs = []
for k in range(q0.shape[0]):
    a = find_alpha2(q0[k], q1[k], w0[k], w1[k])
    q = a*q0[k] + (1-a)*q1[k]
    w = (1-a)*w0[k] + a*w1[k]
    rangs.append((np.linalg.norm(q-q0[k])**2)/n + (np.linalg.norm(w-w0[k])**2)/m)
```

```python
plt.bar(list(range(len(rangs))), rangs)
plt.xlabel('НОМЕР ЭКСПЕРТА')
plt.ylabel('РАНГ')
plt.show()
```

```python
print('Номер лучшего эксперта =', rangs.index(max(rangs)))
```
