```python
import csv
import numpy as np
import pandas as pd
import scipy.spatial.distance
data = []
data = pd.read_csv("Food rating - Sheet1.csv",header = None)
data = data.loc[:12, 0:18]
```

```python
data1 = data.fillna(1000)
data1 = data1.loc[:,2:18]
```

```python
avg = np.array(data1[data1 < 1000].sum())
```

```python
avg = avg/([9,9,9,9,9,7,8,8,7,9,8,8,8,7,7,7,9])
```

```python
avg
```

```python
for i in range(17):
    for j in range(9):
        if(data1.iloc[j,i] == 1000):
            data1.iloc[j,i] = avg[i]
```

```python
data1
```

```python
data2 = np.array(data1).astype(np.float)
```

```python
median = np.mean(data2,axis = 0)
for i in range (17):
    data2[:,i] = data2[:,i] / np.mean(data2[:,i])
```

```python
median
```

```python
for i in range(12):
    avg[5+i] = 1
ideal = avg/median
ideal
```

```python
results = []
for degree in range (11):
    currentresults = []
    for i in range(9):
        currentresults.append( (np.abs(data2[i] - ideal)**(2**(degree - 5))).sum()**(1/(2**(degree - 5))))
    results.append(currentresults)
```

```python
resultsdata = pd.DataFrame((np.array(results).transpose()/np.mean(results,axis = 1 )).transpose())
```

```python
ans = np.zeros([12])
for k in range (12):
    rightrating = [8,9,5,6,7,1,2,4,3]
    myrating = data.fillna(1000).loc[:,k+7]
    myrating = np.array(myrating).astype(np.int)
    for i in range (9):
        for j in range(9):
            ans[k] = ans[k] + (0 < (myrating[i] - myrating[j]) * (rightrating[i] - rightrating[j]) < 1000)
```

```python
finalans = ans/np.array([7,8,8,7,9,8,8,8,7,7,7,9])**2
```
